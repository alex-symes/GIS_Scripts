## GIS LineString Geometry to Lat/Longs

This script pulls the lat/longs from a LineString. This is specifically for EIA map data from: https://www.eia.gov/maps/layer_info-m.php


```python
from shapely.geometry import Point

import geopandas as gpd
import pandas as pd

# open the file and print a lat long.
soa_shape_map_path = "./NaturalGas_InterIntrastate_Pipelines_US_EIA/NaturalGas_Pipelines_US_202001.shp"
soa_shape_map = gpd.read_file(soa_shape_map_path)

soa_shape_map_geo = soa_shape_map.to_crs(epsg=4326)  # EPSG 4326 = WGS84 = https://epsg.io/4326
print(soa_shape_map.iloc[0].geometry.centroid.y, soa_shape_map_geo.iloc[0].geometry.centroid.y)
```

```python
soa_shape_map_geo.columns
```

```python
i = 0
j = 0
cols=['ID_pipe','ID_geo','Typepipe','Operator','Status','Shape Leng','Long','Lat']

df_final = pd.DataFrame(columns=cols)

for index, row in soa_shape_map_geo.iterrows():

    ty = row['TYPEPIPE']
    op = row['Operator']
    st = row['Status']
    sh = row['Shape_Leng']
    ge = row['geometry']
    try: 
        for pt in list(row['geometry'].coords):
            #print([i,j,pt[0],pt[1]])
            df_temp = pd.DataFrame([[i,j,ty,op,st,sh,pt[0],pt[1]]],columns=cols)
            df_final = pd.concat([df_final,df_temp])
            j = j+1
    except:
        pass
    i = i + 1
```
